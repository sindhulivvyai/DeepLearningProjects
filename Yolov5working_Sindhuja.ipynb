```python
# Step 1: Clone YOLOv5 repo and install dependencies
!git clone https://github.com/ultralytics/yolov5
%cd yolov5
%pip install -r requirements.txt

# Step 2: Import required libraries
import torch
from matplotlib import pyplot as plt
import cv2
from PIL import Image

# Step 3: Load a pretrained YOLOv5 model (YOLOv5s is smallest & fastest)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Step 4: Upload an image
from google.colab import files
uploaded = files.upload()

# Step 5: Run detection
for img_path in uploaded.keys():
    results = model(img_path)
    results.print()  # Show results in text
    results.show()   # Show results with bounding boxes (in a popup window)

    # Optional: save result image with boxes
    results.save()

# Step 6: Display result image inline (Optional)
import os
from IPython.display import Image as IPyImage
result_img_path = os.path.join('runs', 'detect', os.listdir('runs/detect')[-1], img_path)
IPyImage(filename=result_img_path)
```
